```python
from matplotlib.patches import Rectangle
import yaml

img = task_manager.controller.screenshot()
# img_resized = cv2.resize(img, (406, 900))
img_resized = img

click_point = (100, 160)

file_name = "Zdy_right_config"
picture_path = f"images/{file_name}.png"
config_path = f"images/{file_name}_config.yaml"
left = 26+142+8
top = 416
width = 142
height = 178
region = {
          "left": left , # 左边界X坐标
          "top": top , # 上边界Y坐标
          "right": left+width ,# 右边界X坐标
          "bottom": top+height ,# 下边界Y坐标 
          "defult_location":{
            "x":click_point[0],
            "y":click_point[1],
            },
          }


fig, ax = plt.subplots()
ax.imshow(img)  # 显示背景图像
width = region['right'] - region['left']
height = region['bottom'] - region['top']
rect = Rectangle((region['left'], region['top']), width, height, linewidth=0.5, edgecolor='red', facecolor='none')
ax.add_patch(rect)
# 在图像上添加圆形标记
circle = plt.Circle(click_point, 5, color='blue', fill=True)
ax.add_patch(circle)

plt.show()
plt.imshow(img_resized)
plt.show()
cropped_img = img_resized[region['top']:region['bottom'], region['left']:region['right']]
# Ask the user if they want to save the cropped image
save_image = input(f"Do you want to save {picture_path}? (yes/no): ").strip().lower()
if save_image == 'yes':
    cv2.imwrite(picture_path, cropped_img)
    # Save the region dictionary to a YAML config file
    with open(config_path, 'w') as file:
        yaml.dump(region, file, default_flow_style=False)
plt.imshow(cropped_img)
plt.show()
task_manager.check_image(
    picture_path,
    region
)
template = cropped_img
result = cv2.matchTemplate(img_resized, template, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
log(f"最佳匹配度: {max_val}, 位置: {max_loc}")
# task_manager.controller.tap(click_point[0], click_point[1])

OCR_region = {
            "left": 68 , # 左边界X坐标
            "top": 140 , # 上边界Y坐标
            "right": 125 ,# 右边界X坐标
            "bottom": 160 ,# 下边界Y坐标 
                }

OCR_img = cropped_img[OCR_region['top']:OCR_region['bottom'], OCR_region['left']:OCR_region['right']]
plt.imshow(OCR_img)
plt.show()
# 转换为灰度图
gray = cv2.cvtColor(OCR_img, cv2.COLOR_BGR2GRAY)

# 二值化处理，使用自适应阈值以提高文字识别率
binary = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    cv2.THRESH_BINARY, 11, 2
)

# 使用OCR识别文字，配置为识别数字和冒号
text = pytesseract.image_to_string(
    binary,
    config='--psm 7 -c tessedit_char_whitelist=+0123456789:'
).strip()
print(f"识别结果: {text}")
```
